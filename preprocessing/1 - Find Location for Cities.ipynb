```python
with open('cities.txt','r') as f:
    data = f.read()
```

```python
data.split('\n')
```

```python
data_loaded = []
for city in data.split('\n'):
    print city
    url = "https://maps.googleapis.com/maps/api/place/textsearch/json?query="+city.replace(' ','%20')
    url += '&key=' + key
    response = urllib2.urlopen(url)
    data_loaded.append(json.load(response))
    
```

```python
import pickle
pickle.dump( data_loaded, open( "data/save.p", "wb" ) )
```

```python
output = [] 
for city in data_loaded:
    if 'results' in city and len(city['results'])>0:
        print city['results'][0]['geometry']['location']
        lat = city['results'][0]['geometry']['location']['lat']
        lng = city['results'][0]['geometry']['location']['lng']
        print city['results'][0]['name']
        name = city['results'][0]['name']
        output.append({
            'name':name,
            'lat':lat,
            'lng':lng
        })
```

```python
with open('cities.json', 'w') as outfile:
    json.dump({'cities':output}, outfile)
```
