```python
import json
```

```python
data = ''
with open('cities.json','r') as f:
    data = json.load(f)
data
```

```python
# 8NN (manual)
def distance(x,y):
    try:
        A = (x['lat'],x['lng'])
        B = (y['lat'],y['lng'])
        return vincenty(A,B).km
    except:
        return 10e10
    return 10e10
    
data_with_8NN = []
for city in data['cities']:
#     print(city)
    new_array = sorted(data['cities'], key=lambda x:distance(x,city))[1:9]
    same_city = city.copy()
    same_city['8NN'] = new_array
    data_with_8NN.append(same_city)
```

```python
data_with_8NN[10]
```

```python
links = []
for city in data_with_8NN:
    for city_connected in city['8NN']:
        links.append({
            "source":{x: city[x] for x in ['name','lat','lng'] },
            "target":city_connected,
            "value":1
        })

added_manually = [
    ('Tripoli','Kano')
    ('Khartoum','Kano')
    ('Khartoum','Brazzaville')
]
```

```python
with open('cities_links.json', 'w') as outfile:
    json.dump({
        'cities':data['cities'],
        'links':links
    }, outfile)
```
